```python
cars2010 = pd.read_csv('./data/cars2010.csv')
cars2011 = pd.read_csv('./data/cars2011.csv')
cars2012 = pd.read_csv('./data/cars2012.csv')
```

```python
a = plt.scatter(cars2010.FE, cars2010.EngDispl, c= 'green')
plt.scatter(cars2011.FE, cars2011.EngDispl, c= 'red')
plt.scatter(cars2012.FE, cars2012.EngDispl, c= 'blue',)

plt.xlabel('Fuel Economy')
plt.ylabel('Engine Displacement')

plt.text(x=50, y=8, s='2010 Cars in Green')
plt.text(x=50, y=7.5, s='2011 Cars in Red')
plt.text(x=50, y=7, s='2012 Cars in Blue')

plt.title('Engine Displacement (l) vs Fuel Economy (mpg)')

plt.show()
```
